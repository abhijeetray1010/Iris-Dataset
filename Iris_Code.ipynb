### Building Model

##### Steps To Be followed When Applying an Algorithm
1)Split the dataset into training and testing dataset. The testing dataset is generally smaller than training one as it will help in training the model better.

2)Select any algorithm based on the problem (classification or regression) whatever you feel may be good.

3)Then pass the training dataset to the algorithm to train it. We use the .fit() method

4)Then pass the testing data to the trained algorithm to predict the outcome. We use the .predict() method.

5)We then check the accuracy by passing the predicted outcome and the actual output to the model.

#### Specifying target and features

```python
target = df['Species']
features = df.drop(columns = ['Species'])
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
target = encoder.fit_transform(target)
```

```python
target
```

#### As you can see Iris-setosa, Iris-versicolor and Iris-virginica are converted to 0, 1, 2 respectively

#### Splitting the dataset into training and testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features, target, train_size=0.7, random_state=42)
```

```python
print('Length of training data',len(X_train))
print('Length of testing data', len(X_test))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

LR_model = LogisticRegression()
LR_model.fit(X_train, y_train) 
# model.score automatically do the predictions(y_predictions) for X_test and then perform calculations between y_train 
# and y_predicted values to find the model score, so no need to need to find y_predicted seperately. if required we can do it
#y_predicted = LR_model.predict(X_test)
LR_model_score = LR_model.score(X_test,y_test)
print(f'The model score is: {LR_model_score}')
```

##### Another way of finding accuracy

```python
from sklearn import metrics 
y_predicted = LR_model.predict(X_test)
print('The accuracy of the Logistic Regression is',metrics.accuracy_score(y_test,y_predicted))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_predicted))
```

### Confusuion Matrix for Logistic Regression

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_predicted)
```

```python
y_predicted
```

```python
cm = confusion_matrix(y_test, y_predicted)
sns.heatmap(cm, annot=True).set_title('Confusion Matrix for Logistic Regression', color='black', size = '15')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT_model = DecisionTreeClassifier()
DT_model.fit(X_train,y_train)
DT_model_score = DT_model.score(X_test, y_test)
print('The model score is',DT_model_score)
```

```python
from sklearn import metrics
y_predicted_dt = DT_model.predict(X_test)
print('The accuracy is', metrics.accuracy_score(y_test, y_predicted_dt))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_predicted_dt))
```

### RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RF_model = RandomForestClassifier(n_estimators=5,max_depth=2,random_state=0)
RF_model.fit(X_train,y_train)
RF_model_score = RF_model.score(X_test,y_test)
print('The model score is', RF_model_score)
```

```python
from sklearn import metrics 

y_predicted_rf = RF_model.predict(X_test)
print('The accuracy is:', metrics.accuracy_score(y_test, y_predicted_rf))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_predicted_rf))
```

### SVM

```python
from sklearn.svm import SVC

svc_model = SVC(C=1.0, kernel = 'rbf',gamma='auto')
svc_model.fit(X_train, y_train)
svc_model_score = svc_model.score(X_test, y_test)
print(svc_model_score)
```

##### We can tune the parameters in case of SVC

```python
from sklearn.svm import SVC

svc_model = SVC(C=10, kernel = 'linear',gamma='auto')
svc_model.fit(X_train, y_train)
svc_model_score = svc_model.score(X_test, y_test)
print(svc_model_score)
```

```python
# We can tune the values of C(Regularization, Kernel and gamma) to increase our model performance if required, 
#but in our case, tuning is not required the default values of C, Kernel and gamma gave us 1.0 score
```

```python
from sklearn import metrics
y_predicted_svc = svc_model.predict(X_test)
print('The accuracy is:', metrics.accuracy_score(y_test, y_predicted_svc) )
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_predicted_svc))
```

### XGBOOST

```python
import xgboost as xgb

xgb_model = xgb.XGBClassifier()
xgb_model.fit(X_train, y_train)
xgb_model_score = xgb_model.score(X_test, y_test)
print(svc_model_score)
```

##### We can tune the parameters in case of SVC

```python
train= xgb.DMatrix(X_train, label=y_train)
test= xgb.DMatrix(X_test, label=y_test)
```

```python
param = {
    'max_depth':4,
    'eta': 0.3,
    'objective': 'multi:softmax',
    'num_class':3
}
epochs = 10
```

```python
model=xgb.train(param, train, epochs) 
```

```python
predictions=model.predict(test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, predictions)
```

### Lets use K-Fold Cross validation technqiue 

### cross_val_score function

```python
from sklearn.model_selection import cross_val_score
```

#### K-Fold on Logistic Regression

```python
l_score=cross_val_score(LogisticRegression(solver='liblinear',multi_class='ovr'), features,target, cv=10)
l_score
```

```python
np.average(l_score)
```

#### K-Fold on Decision Tree

```python
dt_score = cross_val_score(DecisionTreeClassifier(), features,target, cv=10)
dt_score
```

```python
np.average(dt_score)
```

##### Just for info, we can also do parameter tuning using K-fold cross validation on other models to check if they can be improved

```python
# Paramter tuning on Decision Tree
dt_1_score = cross_val_score(DecisionTreeClassifier(max_depth=5), features,target, cv=10)
np.average(dt_1_score)
```

```python
dt_2_score = cross_val_score(DecisionTreeClassifier(max_depth=10), features,target, cv=10)
np.average(dt_2_score)
```

```python
dt_3_score = cross_val_score(DecisionTreeClassifier(max_depth=100), features,target, cv=10)
np.average(dt_3_score)
```

##### No Change in the performance for Decision Tree

#### K-Fold on Random Forest Classifier

```python
rf_score = cross_val_score(RandomForestClassifier(n_estimators=40), features, target, cv=10)
rf_score
```

```python
np.average(rf_score)
```

```python
#Parameter tuning for Random Forest Classifier
rf_1_score = cross_val_score(RandomForestClassifier(n_estimators=100), features, target, cv=10)
np.average(rf_1_score)
```

```python
rf_2_score = cross_val_score(RandomForestClassifier(n_estimators=150), features, target, cv=10)
np.average(rf_2_score)
```
